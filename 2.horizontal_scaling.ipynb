# OpenDC Demo 2
### Horizontal Scaling

In the [experiment 1](1.first_experiment.ipynb), we learned how to run OpenDC experiments, and how to analyze and visualize the results to learn about the behavior of a data center. In this demo we are using OpenDC to determine the impact of horizontal scaling. Horizontal scaling refers to reducing (or increasing) the number of machines available in a data center to better fit a workload.

Although it is always good to have a lot of machines for performance, having idle machines can result in a significant energy cost. On the other hand, having too few machines can significantly reduce the performance of the data center. In this demo we are running the same workload (surf_week) on four different data center scales. After, we analyse the results to get an understanding of the impact of data center size.

## Output

#### Exercise 3: 
Load the results into Pandas DataFrames

```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt

df_host_100 = pd.read_parquet("output/2.horizontal_scaling/raw-output/0/seed=0/host.parquet")
df_powerSource_100 = pd.read_parquet("output/2.horizontal_scaling/raw-output/0/seed=0/powerSource.parquet")
df_task_100 = pd.read_parquet("output/2.horizontal_scaling/raw-output/0/seed=0/task.parquet")
df_service_100 = pd.read_parquet("output/2.horizontal_scaling/raw-output/0/seed=0/service.parquet")

df_host_150 = pd.read_parquet("output/2.horizontal_scaling/raw-output/1/seed=0/host.parquet")
df_powerSource_150 = pd.read_parquet("output/2.horizontal_scaling/raw-output/1/seed=0/powerSource.parquet")
df_task_150 = pd.read_parquet("output/2.horizontal_scaling/raw-output/1/seed=0/task.parquet")
df_service_150 = pd.read_parquet("output/2.horizontal_scaling/raw-output/1/seed=0/service.parquet")

df_host_200 = pd.read_parquet("output/2.horizontal_scaling/raw-output/2/seed=0/host.parquet")
df_powerSource_200 = pd.read_parquet("output/2.horizontal_scaling/raw-output/2/seed=0/powerSource.parquet")
df_task_200 = pd.read_parquet("output/2.horizontal_scaling/raw-output/2/seed=0/task.parquet")
df_service_200 = pd.read_parquet("output/2.horizontal_scaling/raw-output/2/seed=0/service.parquet")

df_host_279 = pd.read_parquet("output/2.horizontal_scaling/raw-output/3/seed=0/host.parquet")
df_powerSource_279 = pd.read_parquet("output/2.horizontal_scaling/raw-output/3/seed=0/powerSource.parquet")
df_task_279 = pd.read_parquet("output/2.horizontal_scaling/raw-output/3/seed=0/task.parquet")
df_service_279 = pd.read_parquet("output/2.horizontal_scaling/raw-output/3/seed=0/service.parquet")
```

## Aggregating results

#### Exercise 4: 
Print the runtime and utilization of each of the datacenters.

```python
runtime_100 = pd.to_timedelta(df_service_100.timestamp.max() - df_service_100.timestamp.min(), unit="ms")
runtime_150 = pd.to_timedelta(df_service_150.timestamp.max() - df_service_150.timestamp.min(), unit="ms")
runtime_200 = pd.to_timedelta(df_service_200.timestamp.max() - df_service_200.timestamp.min(), unit="ms")
runtime_279 = pd.to_timedelta(df_service_279.timestamp.max() - df_service_279.timestamp.min(), unit="ms")

print(f"The datacenter with 100 hosts finished the workload in {runtime_100}")
print(f"The datacenter with 150 hosts finished the workload in {runtime_150}")
print(f"The datacenter with 200 hosts finished the workload in {runtime_200}")
print(f"The datacenter with 279 hosts finished the workload in {runtime_279}")

utilization_100 = df_host_100.cpu_utilization.mean()
utilization_150 = df_host_150.cpu_utilization.mean()
utilization_200 = df_host_200.cpu_utilization.mean()
utilization_279 = df_host_279.cpu_utilization.mean()

print(f"On average, the utilization of each host in the datacenter with 100 hosts is {utilization_100 * 100:.2f}%")
print(f"On average, the utilization of each host in the datacenter with 150 hosts is {utilization_150 * 100:.2f}%")
print(f"On average, the utilization of each host in the datacenter with 200 hosts is {utilization_200 * 100:.2f}%")
print(f"On average, the utilization of each host in the datacenter with 279 hosts is {utilization_279 * 100:.2f}%")
```

### Sustainability

#### Exercise 4: 
Print the total energy usage and carbon emissions of the four data centers.

```python
energy_100 = df_powerSource_100.energy_usage.sum() / 3_600_000 # convert energy to kWh
energy_150 = df_powerSource_150.energy_usage.sum() / 3_600_000 # convert energy to kWh
energy_200 = df_powerSource_200.energy_usage.sum() / 3_600_000 # convert energy to kWh
energy_279 = df_powerSource_279.energy_usage.sum() / 3_600_000 # convert energy to kWh


print(f"The datacenter with 100 hosts used {energy_100:.2f} kWh during the workload")
print(f"The datacenter with 150 hosts used {energy_150:.2f} kWh during the workload")
print(f"The datacenter with 200 hosts used {energy_200:.2f} kWh during the workload")
print(f"The datacenter with 279 hosts used {energy_279:.2f} kWh during the workload")

carbon_100 = df_powerSource_100.carbon_emission.sum() / 1000 # convert carbon to kg
carbon_150 = df_powerSource_150.carbon_emission.sum() / 1000 # convert carbon to kg
carbon_200 = df_powerSource_200.carbon_emission.sum() / 1000 # convert carbon to kg
carbon_279 = df_powerSource_279.carbon_emission.sum() / 1000 # convert carbon to kg

print(f"The datacenter with 100 hosts emitted {carbon_100:.2f} kg during the workload")
print(f"The datacenter with 150 hosts emitted {carbon_150:.2f} kg during the workload")
print(f"The datacenter with 200 hosts emitted {carbon_200:.2f} kg during the workload")
print(f"The datacenter with 279 hosts emitted {carbon_279:.2f} kg during the workload")
```

## Visualization

#### Exercise 5: 
Plot the number of active tasks during the workload for each of the data center

```python
plt.plot(df_service_100.tasks_active, label="100")
plt.plot(df_service_150.tasks_active, label="150")
plt.plot(df_service_200.tasks_active, label="200")
plt.plot(df_service_279.tasks_active, label="279")

plt.title("Active tasks during a workload")
plt.xlabel("time (h)")
plt.ylabel("active tasks")
plt.legend()
plt.show()
```
